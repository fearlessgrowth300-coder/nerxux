### 4. Talk to your finished model
It should now write real, working code — because it's a full-scale brain, tuned on your data.

```python
from transformers import pipeline
chat = pipeline('text-generation', model='nexus-finetuned-merged', device_map='auto')
msg = [{'role':'user','content':'Write a Python function that checks if a string is a palindrome.'}]
print(chat(msg, max_new_tokens=200)[0]['generated_text'][-1]['content'])
```
